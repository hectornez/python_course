```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
from scipy import stats
```

```python
tips = sns.load_dataset('tips')
```

# .barplot()
### It is a general plot that allows u to asses a categorical column based on a function, which default is the mean

```python
## this can be understood as the visualization of a group by action

sns.barplot(x='sex', y='total_bill', data=tips)

## sns.barplot(x='sex', y='total_bill', data=tips, estimator=np.std)
```

# .countplot()

```python
## this essentially counts how many 
sns.countplot(x='sex', data=tips)
```

# .boxplot()

```python
#sns.boxplot(x='day', y='total_bill', data = tips)
# Este diagrama de cajas muestran los 'mosaicos centrales del conjunto de datos', mientras que los extremos o bigotes
#### se extienden para mostrar el resto de la distribución
## los valores fuera de estos bigotes no son representativos, son casos aislados

### Pointing at Thursday, we see that about 50% is located in the inner box, which is separated in two parts
## of 25% each one. then we have other 25% below the box, and other 25% above it

sns.boxplot(x='day', y='total_bill', data = tips, hue='smoker') #we can add other category by hue='column name'
```

# .violinplot()

```python
## we can observe a white spot in each diagram which show the middle point. The distribution
### is essential the same as the boxplot. below the white spot we have the 20% till the black thick line ends, 
## then begins the other 25% which is the thin black line. And the same upwards.
### Understanding the box, it becomes straightforward to get it

sns.violinplot(x='day', y='total_bill', data=tips, hue='smoker', split=True)
```

# .stripplot()
### Shows and scatter plot, where one variable is categorical

```python
sns.stripplot(x = 'day', y='total_bill', data=tips, jitter=True, hue='smoker', split=True) 
         ## Jitter adds some random noise## for better visualization
```

# .swarmplot()
### It resembles as a combination of .stripplot() and .violinplot()

```python
sns.swarmplot(x='day', y='total_bill', data=tips)
```

```python
sns.swarmplot(x='day', y='total_bill', data=tips, color='y')
sns.violinplot(x='day', y='total_bill', data=tips)
```

# .factorplot()
## by kind=''     we can create each of the ones alreay explained

```python
sns.factorplot(x='day', y='total_bill', data=tips, kind='bar')
```

```python
sns.factorplot(x='day', y='total_bill', data=tips, kind='violin')
```
