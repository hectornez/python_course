# Now we begin to create our model

```python
from sklearn.model_selection import train_test_split
```

```python
df.info()
```

First we split the data

```python
X = df.drop('benign_0__mal_1', axis=1).values
y = df['benign_0__mal_1'].values
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=101)
```

Now we scale our data

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
scaler = MinMaxScaler()
```

```python
X_train = scaler.fit_transform(X_train)
```

```python
X_test = scaler.transform(X_test)
```

NOw we create our model

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense,Dropout
```

```python
X_train.shape
```

```python
model = Sequential()

model.add(Dense(30, activation='relu'))
model.add(Dense(15, activation='relu'))
#For binary classification model de last layer must have a sigmoid function as activation function
model.add(Dense(1, activation='sigmoid'))

model.compile(loss='binary_crossentropy',optimizer='adam')
```

Now we fit our model with the train data

```python
model.fit(x=X_train, y=y_train, epochs=600, 
         validation_data=(X_test, y_test))
```

```python
losses=pd.DataFrame(model.history.history)
```

```python
losses.plot()
```

After ploting our losses we observe that are overfitting. Essentially our loss and val_loss are both decreasing, but at a certain point our val_loss begins to rise. So it means that we are overfitting to training data. 

NOW we will see if we can use EARLY STOPPING, beacuse we obviously have trained too much.

Then we copy and paste our model again

```python
model = Sequential()

model.add(Dense(30, activation='relu'))
model.add(Dense(15, activation='relu'))
#For binary classification model de last layer must have a sigmoid function as activation function
model.add(Dense(1, activation='sigmoid'))

model.compile(loss='binary_crossentropy',optimizer='adam')
```

```python
from tensorflow.keras.callbacks import EarlyStopping
```

We are going to use Early Stopping and there are some tips about it:

The 'monitor' is essentially what we are trying to control and then stop before it gets the critical point. Critical point for the loss is when it begins to increase instead of decrease.

The 'mode' we are using can be 'auto', 'min' or 'max'. For example, if we are monitoring loss we always want to minimize it(so we choose 'min'), but if we are monitoring  accuray, we will want to maximize it(so we get 'max'). In addition, 'auto' will be suitable for the machine do it automatically, based on the monitor name.

The 'patience' is the number of epochs with no improvement after which training will be stopped. We can set a certain number, due to some noise can raise and after that our loss can keep decreasing.

```python
early_stop = EarlyStopping(monitor='val_loss', mode='min', verbose=1, patience=25)
```

Now we have to fit our model again. Our model will try to carry out 600 epochs, but the earlystopping will be triggered when necessary.

With this method we don 't have to worry about how large our epochs number must be

```python
model.fit(x=X_train, y=y_train, epochs=600, 
         validation_data=(X_test, y_test),
         callbacks=[early_stop])#we pass callbacks as a list []
```

```python
losses_early_stop = pd.DataFrame(model.history.history)
losses_early_stop.plot()
```

This exactly how it should look like, since what we want to prevent is for the val_loss go raise under control

Ohter thing that we can do to prevent overfitting is to use/add DROPOUT LAYERS. This will essentially turn off some neurons (a %) randomly.

W will copy and paste our model again

```python
from tensorflow.keras.layers import Dropout
```

Right below every layer (model.add) we will add our Dropout

we must especify a certain percentage of neurons that are going to be turned of randomly. It means that this % during each batch. SInce we have not especified the batch size, it will be one entire epoch of training.

Half of the neurons of the first layer will turn off(randomly), so the weights and biases won't be updated.
We repeat the same operation in the second layer.

```python
model = Sequential()

model.add(Dense(30, activation='relu'))
model.add(Dropout(0.5))

model.add(Dense(15, activation='relu'))
model.add(Dropout(0.5))

#For binary classification model de last layer must have a sigmoid function as activation function
model.add(Dense(1, activation='sigmoid'))

model.compile(loss='binary_crossentropy',optimizer='adam')
```

Finally we add our earlystoping (callbacks) as well. Both the early stoping and the Dropout should help to reduce overfitting

```python
model.fit(x=X_train, y=y_train, epochs=600, 
         validation_data=(X_test, y_test),
         callbacks=[early_stop])#we pass callbacks as a list []
```

We realize that our early stopping was triggered later than before, and this is a good sign of the model to be learning after those further epochs  due to the dropout layers

```python
losses_early_stop_dropout = pd.DataFrame(model.history.history)
```

```python
losses_early_stop_dropout.plot()
```

Eventually we have reduce greatly our overrfitting.

We recall that we are performing a clasiffication task, SO WE ARE PREDICTING A CLASS 0 OR 1.

Since .predict_classes is deprecated we have to use .predict which will return a continious value. Then we will turn these values into 0 and 1 based on if they are above or below 0.5

Then we can evaluate our model

```python
predictions = model.predict(X_test)
```

```python
predictions
```

```python
predicted_classes = predictions>0.5
```

```python
predicted_classes = predicted_classes.astype(int)
```

```python
predicted_classes
```

```python
from sklearn.metrics import confusion_matrix,classification_report
```

```python
print(classification_report(y_test, predicted_classes))
```

```python
print(confusion_matrix(y_test,predicted_classes))
```
