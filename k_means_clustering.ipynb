```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
from sklearn.datasets import make_blobs
```

```python
data = make_blobs(n_samples=200, n_features=2, centers=4, cluster_std=1.8,
                 random_state=101)
```

```python
data
```

data is a tupple so it has two blocks of data [0] and [1], as you can observe above. For calling them in a scatter plot it must be done as follows

```python

plt.scatter(x=data[0][:,0], y=data[0][:,1], c=data[1] )
```

```python
from sklearn.cluster import KMeans
```

```python
kmeans = KMeans(n_clusters=4)
```

```python
kmeans.fit(data[0])
```

```python
kmeans.cluster_centers_
```

```python
kmeans.labels_
```

```python
fig, (ax1,ax2) = plt.subplots(1,2, sharey=True)

ax1.set_title('K_means')
ax1.scatter(data[0][:,0], data[0][:,1], c=kmeans.labels_)

ax2.set_title('Original dat')
ax2.scatter(data[0][:,0], data[0][:,1], c=data[1])
```
