```python
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline   ### this allows to see the plots in jupyter without using plt.show()
```

```python
x = np.linspace(0,5,11)
y = x**2
```

```python
plt.plot(x,y)
#plt.show()
```

```python
plt.subplot(1,2,2) ### Here we say - plt.subplot(number of rows, number of columns, plot number u are refering to)
plt.plot(x,y,'r')


plt.subplot(1,2,1)
plt.plot(y,x,'b')
```

```python
fig = plt.figure()  #This creates a blank figure where we can have a bit more control over it

axes1 = fig.add_axes([0.1,0.5,0.9,0.9]) ## Here we set the figure parameters(position and length of the plot) 
                                        # in percentage , so numbers in the  ## must be 0-1 .
                                        
axes2 = fig.add_axes([0.2,0.8,0.3,0.3])    
 
axes1.plot(x,y)
axes1.set_title('exponential')
axes2.plot(y,x)
axes2.set_title('logarithm')

```

```python
## The difference with plt.figure( is that plt.subplots() makes this (axes = fig.add_axes([0.1,0.5,0.9,0.9])) 
    ## automatically for us, and creates as many as we say. for example here we create 9
    ## figsize is in inches, and dpi is the resolution que want
fig,axes = plt.subplots(nrows=1,ncols=2, figsize=(9,5), dpi=100)
plt.tight_layout()

axes[0].plot(x,y)
axes[0].set_title('plot 1')
axes[1].plot(y,x)
axes[1].set_title('plot 2')
# axes is essentially an array, so we can iterate with it
```
