# .heatmap
### In order to this function works properly, ourdata must be in a matrix form. It means that each col or row, must be under a variable or label. So, looking at the flights data set, we observe that 'year', 'month' and 'passengers' are the labels or variables, but the rows, where we can see the index are not acctually.
### We can transform it by using pivot table, or correlation data.

```python
fl = flights.pivot_table(index='month', columns='year', values = 'passengers')
```

```python
fl
```

```python
sns.heatmap(fl, cmap='magma', linecolor='white', linewidths=1)
```

```python
## .clustermap() arrange the cols and rows according to the resemblance from one another.
## So we can observe that 1959 and 1960 are quite similar years. Having a look to the hierrachi scheme
### we can see which years and months ended up with some kind of resemblance.
sns.clustermap(fl) 
## sns.clustermap(fl, standard_scale=1) 
```
