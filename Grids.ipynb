```python
import seaborn as sns
import matplotlib.pyplot as plt
iris = sns.load_dataset('iris')
iris.head()
```

```python
iris['species'].unique()
```

```python
sns.pairplot(iris)
```

```python
# As we can observe .PairGrid() seems as .pairplot(), and it is owing to .pairplot() comes from .PairGrid()
## This means that with .PairGrid(), wee can create the .pairplot() we want
g = sns.PairGrid(iris)
```

```python
g = sns.PairGrid(iris)
g.map_diag(plt.hist)
g.map_upper(plt.scatter)
g.map_lower(sns.kdeplot)
```
